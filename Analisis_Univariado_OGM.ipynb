# Descripcion del Conjunto de Datos

El conjunto de datos que analizaremos corresponde a uno tomado de Kaggle, denominado "Domestic Violence in Colombia" (Violencia Domestica en Colombia). Como el conjunto de datos incluye casi 500 mil registros. Decidi tomar solo los 5 mil primeros registros para el analisis univariado.

Estos datos fueron tomados entre los años 2010 a 2021.

El enlace al conjunto de datos es el siguiente:
https://www.kaggle.com/datasets/oscardavidperilla/domestic-violence-in-colombia/code

```python
# Iniciamos importando la librería de Pandas, que nos permite cargar el conjunto de datos (dataset) que vamos
# a usar y analizar.
import pandas as pd

# Cargamos el conjunto de datos que vamos a analizar. Recordemos que fue tomado de kaggle y corresponde

url = 'https://github.com/octagm/local_ia_2024/blob/main/Reporte_Delito_Violencia_Intrafamiliar_Polic_a_Nacional.csv?raw=true'
df = pd.read_csv(url)
```

# Visualización de los Datos

Vamos a utilizar las librerías de Python: Matplotlib y seaborn para graficar los registros del conjunto de datos cargado "Domestic Violence in Colombia". Realizamos el analsis univariado sobre la variable "CANTIDAD", que refiere el número de casos de violencia reportados.

```python
# Primero Importamos las librerias Matplotlib y seaborn para graficar
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Iniciemos graficando la cantidad de Incidencias en Casos de Violencia(CANTIDAD) basado en el numero de registros en el conjunto de datos
plt.figure(figsize=(10,10))
plt.scatter(df.index,df['CANTIDAD'])
plt.show()
```

# Análisis de la Gráfica generada (plt.scatter)
Del grafico anterior podemos inferir:

1. La mayoría del número de casos de Violencia estan en el rango de los 0 a 5.
2. Un numero no menor esta en el rango de 5 a 10.
3. En el rango de 10 o mas, hay valores, pero definitivamente, no son la mayoría.

```python
# Evaluemos la grafica anterior pero incluyamos el tipo de ARMA o MEDIO que se usaron en los casos de violencia reportados.
plt.figure(figsize=(15,15))
sns.scatterplot(x=df.index,y=df['CANTIDAD'],hue=df['ARMAS MEDIOS'])
plt.show()
```

# Análisis de la Gráfica
Podemos notar que los casos de Violencia Domestica en su mayoria han sido realizados con armas contundentes, que es el valor mas reflejado en ella.

```python
# Realicemos un Diagrama de Lineas del Numero de casos agrupado por Armas o Medios usados en ellos.
plt.figure(figsize=(15,15))
plt.title('Diagrama de Lineas para Nro. de Casos por Armas y Medios')
plt.xlabel('index',fontsize=20)
plt.ylabel('Nro. de Casos',fontsize=20)
plt.plot(df.index,df['CANTIDAD'],markevery=1,marker='d')
for name, group in df.groupby('ARMAS MEDIOS'):
    plt.plot(group.index, group['CANTIDAD'], label=name,markevery=1,marker='d')
plt.legend()
plt.show()
```

# Analisis de la Gráfica
Es notorio de nuevo que en la mayoría de  casos de Violencia, el arma o medio usado es "arma contundente".  

```python
# Hagamos un gráfico de enjambre para la cantidad de casos
sns.set(rc={'figure.figsize':(25,25)})
sns.stripplot(x=df['ARMAS MEDIOS'],y=df['CANTIDAD'])
```

```python
# Generemos un histograma para los casos de violencia
plt.rcParams["figure.figsize"] = (6,6)
plt.hist(df['CANTIDAD'])
```
